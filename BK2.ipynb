```python
import pandas as pd
import re
import numpy as np
import itertools
```

load data

```python
dir = 'hungarian.data'
```

membuka file yang diberikan oleh path yang disimpan dalam variabel 'dir' dan membaca baris-baris di dalamnya.

```python
with open(dir, encoding='Latin1') as file:
  lines = [line.strip() for line in file]
lines[0:10] #menggunakan slicing untuk mengambil sepuluh elemen pertama dari list 'lines'. Hasilnya adalah sepuluh baris pertama dari file yang dibaca.
```

membaca dan memproses baris-baris dari file yang telah dibaca sebelumnya, mengambil kelompok-kelompok sepuluh baris dengan panjang total 76, dan membangun DataFrame dari hasilnya untuk kemudian menampilkan lima baris pertama dari DataFrame tersebut.

```python
data = itertools.takewhile(lambda x: len(x) == 76,
    (' '.join(lines[i:(i+10)]).split() for i in range(0, len(lines),10)))
df = pd.DataFrame.from_records(data)
df.head()
```

**7. MODEL**

```python
#Sebuah fungsi untuk menampilkan hasil akurasi dan rata - rata dari recall , f1 dan precision score setiap model

from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score,roc_auc_score,confusion_matrix,precision_score
```

```python
def evaluation(Y_test,Y_pred):
  acc = accuracy_score(Y_test,Y_pred)
  rcl = recall_score(Y_test,Y_pred,average = 'weighted')
  f1 = f1_score(Y_test,Y_pred,average = 'weighted')
  ps = precision_score(Y_test,Y_pred,average = 'weighted')

  metric_dict={'accuracy': round(acc,3),
               'recall': round(rcl,3),
                'F1 score': round(f1,3),
                'Precision score': round(ps,3)
              }

  return print(metric_dict)
```
